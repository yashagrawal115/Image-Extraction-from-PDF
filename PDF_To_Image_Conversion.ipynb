## PDF_To_Image_Conversion

```python
from wand.image import Image as wi
import os
from PyPDF2 import PdfFileReader
from wand.color import Color
from PIL import Image
import Image
from pytesseract import image_to_string

def getImageName(file_path):
    file_name = os.path.basename(file_path)
    index_of_dot = file_name.index('.')
    file_name_without_extension = file_name[:index_of_dot]
    return file_name_without_extension


def getListOfFiles(dirName):
    # create a list of file and sub directories
    # names in the given directory
    listOfFile = os.listdir(dirName)
    allFiles = []

    # Iterate over all the entries
    for entry in listOfFile:
        print("Entry is",entry)
        # Create full path
        fullPath = os.path.join(dirName, entry)
        # If entry is a directory then get the list of files in this directory
        if os.path.isdir(fullPath):
            allFiles = allFiles + getListOfFiles(fullPath)
            print("all files", allFiles)
        else:
            fullPath = fullPath.replace("\\", "/")
            allFiles.append(fullPath)

    return allFiles


def pdfToImageConversion(pdfName, imageCount):
    pdf = wi(filename=pdfName, resolution=300)
    pdfimage = pdf.convert("jpg")

    fileName = os.path.split(pdfName)[1]
    filePath = os.path.split(pdfName)[0]
    # filePath = "//sauAzeOCR01\Demo\Data\gpcimage"
    oldImageName = getImageName(pdfName)
    if not os.path.exists(filePath):
        os.mkdir(filePath)
        print("Directory ", filePath, " Created ")
    else:
        print("Directory ", filePath, " already exists")

    i = 0
    count = 0

    for img in pdfimage.sequence:
        page = wi(image=img)
        imagePath = str(filePath + '/' + oldImageName + '_' + str(i)) + ".jpg"
        # page.resize(800,1034)
        page.save(filename=imagePath)
        print(oldImageName + '_' + str(i) + ".jpg", imagePath,
              '',
              '', '', '', 0)

        img = Image.open(imagePath)  # image extension *.png,*.jpg
        # new_width = 612
        # new_height = 792
        # img = img.resize((new_width, new_height), Image.ANTIALIAS)
        img.save(imagePath)

        textfilename = oldImageName + '_' + str(i)
        extractdata = image_to_string(Image.open(imagePath), config="--psm 6")
        textfile = open(filePath + "\\" + textfilename + '.txt', 'a')
        print(filePath + "\\" + textfilename + '.txt')
        textfile.write(extractdata)
        textfile.close()

        count = count + 1
        i = i + 1

imageCount = 0
listOfFiles = getListOfFiles(r'C:\Users\yash\PycharmProjects\Test\pdfs')
# #
for val in listOfFiles:
    if (val.__contains__('.pdf')):
        print("'" + val + "',")
        imageCount = imageCount + 1
        pdfToImageConversion(val, imageCount)
```
